```python
import json
from arcgis.gis import GIS
gis = GIS("home")

print("Successfully logged into AGOL")
```

```python
#uses pandas module to handle importing the .xlxs information and population of python list
import pandas as pd
#create funtion used to email users created from the .xlxs
def email_user(userList):
    for user in userList:
             #message created for the body of the email 
        message = f'<p><h2>Deletion of ArcGIS Online (AGOL) Account</h2></p>'\
        f'<p>The USFS AGOL system is transitioning to eAuthentication (eAuth) login for all named user accounts.  Your named user account, {user}, has been inactive for over a year, and does not contain any content.  If you want to maintain your group membership in a new eAuth account, please send me an email and I can help you migrate your account by Thursday, November 10th. Otherwise your account will be deleted.</p>'\
        '<p>***This is an autogenerated email, please do not hit the reply button, instead forward the email to the Region 9 WebGIS Administrator***</p>'\
        f'<p>Please contact the Region 9 WebGIS Administrator, Robert.Geitner@usda.gov, if you would like your {user} account migrated to an eAuth Account.</p>'\
       #uses python api to email users through the AGOL org
        gis.users.send_notification([user],'USFS Region 9 AGOL Account Information', message, type='email')

        
#get path which includes spreadsheet name to get users list        
r9members = input("Type excel path with filename:")
data = pd.read_excel (r9members, sheet_name='test')
df = pd.DataFrame(data, columns=['Username'])
#print (df)
userList=[]
#loop through spreadsheet to append username to userList
for index, rows in df.iterrows():
    my_list=[rows.Username]
    userList.append(my_list)
#Loop through list and populate function
for user in userList:
    email_user(user)
#print("User notified")

userCount = len(userList)
print("Total Users:",userCount)
print("Users Notified:",userList)
```
